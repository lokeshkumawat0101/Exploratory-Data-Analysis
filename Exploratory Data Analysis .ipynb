```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv('SampleSuperstore.csv')
sns.jointplot(x='Sales',y='Profit',data=df)
```

```python
Segments=df['Segment'].unique()                                  #pie chart of segments
data = df['Segment'].value_counts()/len(df['Segment'])*100
plt.figure(figsize=(12,7))
plt.pie(data,labels=Segments,autopct='%.1f%%')
```

```python
Subcategory=df['Sub-Category'].unique()                     #pie chart of sub categories
data = df['Sub-Category'].value_counts()
plt.figure(figsize=(12,8))
plt.pie(data,labels=Subcategory,autopct='%1.1f%%')
```

```python
category=df['Category'].unique()                       #pie chart of categories
data = df['Category'].value_counts()
plt.figure(figsize=(12,8))
plt.pie(data,labels=category,autopct='%1.1f%%')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df = pd.read_csv('SampleSuperstore.csv')
hm = df.corr()
hm
```

```python
sns.clustermap(hm,cmap = 'coolwarm',annot = True)
```

```python
plt.figure(figsize=(12,8))                        #bar plot of sales and profit in different regions
plt.subplot(1,2,1)                              
sns.barplot(x='Region',y='Sales',data=df)         #Sales in central region is high but profits are relatively low
plt.subplot(1,2,2)
sns.barplot(x='Region',y='Profit',data=df)
```

```python
plt.figure(figsize=(12,8))                              #bar plot of sales and profit in different segments
plt.subplot(1,2,1)
sns.barplot(x = 'Segment', y = 'Sales',data=df)
plt.subplot(1,2,2)
sns.barplot(x = 'Segment', y = 'Profit', data=df)
```

```python
plt.figure(figsize=(12,8))                                  #bar plot of sales and profit by different ship modes
plt.subplot(1,2,1)
sns.barplot(x = 'Ship Mode', y = 'Sales',data=df)
plt.subplot(1,2,2)
sns.barplot(x = 'Ship Mode', y = 'Profit', data=df)
```

```python
plt.figure(figsize=(12,8))                                 #bar plot of sales and profit in different categories
plt.subplot(1,2,1)
sns.barplot(x = 'Category', y = 'Sales',data=df)           #Technology gives highest sales and profit.
plt.subplot(1,2,2)
sns.barplot(x = 'Category', y = 'Profit', data=df)
```

```python
df1 = df[df['Category']=='Furniture']
df2 = df[df['Category']=='Office Supplies']
df3 = df[df['Category']=='Technology']
```

```python
plt.figure(figsize=(12,8))                                #Tables and bookcases inspite of high sales are giving loss 
plt.subplot(1,2,1)
sns.barplot(x = 'Sub-Category', y = 'Sales',data=df1)    #furnishings even with small sales is giving reasonable profits
plt.subplot(1,2,2)
sns.barplot(x = 'Sub-Category', y = 'Profit', data=df1)
```

```python
plt.figure(figsize=(14,10))                                 #Supplies inspite of good sale is giving loss.
plt.subplot(2,1,1)
sns.barplot(x = 'Sub-Category', y = 'Sales',data=df2)
plt.subplot(2,1,2)
sns.barplot(x = 'Sub-Category', y = 'Profit', data=df2)
```

```python
plt.figure(figsize=(12,8))                                #copiers in this subcategory gives highest sales and profit
plt.subplot(1,2,1)
sns.barplot(x = 'Sub-Category', y = 'Sales',data=df3)
plt.subplot(1,2,2)
sns.barplot(x = 'Sub-Category', y = 'Profit', data=df3)
```
